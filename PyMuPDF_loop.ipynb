```python
from pathlib import Path
import fitz
```

```python
pdf_search = Path("/Users/poth/Documents/Fall2022/DEAN690/Coding/TCDS/").glob("*.pdf")
```

```python
pdf_files = pdf_files = [str(file.absolute()) for file in pdf_search]
```

```python
for pdf in pdf_files:
    all_text = ""
    with fitz.open(pdf) as doc:
        for page in doc:
            text = page.getText()
            all_text += '\\n'+ text
```

```python
print(all_text)
```

```python
# OCR the PDF using the 'blocks' parameter
for pdf in pdf_files:
    all_text = ""
    with fitz.open(pdf) as doc:
        for page in doc:
            text = page.get_text()
print(text)
```

```python
for pdf in pdf_files:
    all_text = ""
    with fitz.open(pdf) as doc:
        for page in doc:
            text = page.get_text.encode("utf8")
print(text)
```
