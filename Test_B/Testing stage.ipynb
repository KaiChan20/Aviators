```python

import os
import pdfplumber
import sys
import subprocess
directory = r'./'
for filename in os.listdir(directory):
    if filename.endswith('.pdf'):
        fullpath = os.path.join(directory, filename)
        all_text = ""
        with pdfplumber.open(fullpath) as pdf:
            for page in pdf.pages:
                text = page.extract_text()
                all_text += '\\n' + text
        print(all_text)
```
